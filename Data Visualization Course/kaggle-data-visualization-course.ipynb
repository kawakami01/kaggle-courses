## Scatter Plots

### Set up

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

### Load the data

```python
insurance_data = pd.read_csv(r'C:\Users\migsk\Documents\Kaggle Courses\kaggle-courses\Data Visualization Course\interesting-data-to-visualize\insurance.csv', low_memory=False)
```

```python
insurance_data
```

### Scatter Plots

Para criar um **scatter pllot** simples, nÃ³s usamos o comando *sns.scatterplot*, especificando os valores para:
 - o eixo horizontal (x)
 - o eixo vertical (y)

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges']);
```

O **scatterplot** acima sugere que *body mass index - BMI* (Ã­ndice de massa corporal - IMC) e *insurance charges* (taxa de seguro) sÃ£o <font colro='LimeGreen'>**positivamente correlacionado**</font>, onde clientes com maiores BMI normalmente tambÃ©m tendem a pagar mais em custos de seguro.

> Esse padrÃ£o faz sentido, jÃ¡ que o IMC alto costuma estar associado a um risco maior de doenÃ§as crÃ´nicas

Para uma checagem mais reforÃ§ada nessa relaÃ§Ã£o, vocÃª deve adicionar uma **linha de regressÃ£o**, ou a melhor linha que encaixar nos dados. NÃ³s fazemos isso com o comando *sns.regplot*.

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges']);
```

### Color-codded scatter plots

NÃ³s conseguimos usar o *scatter plot* para mostrar a relaÃ§Ã£o entre trÃªs variÃ¡veis! Uma maneira de fazer isso Ã© codificar os pontos em cores.

Por exemplo, para entender como fumar afeta a relaÃ§Ã£o entre IMC e a taxa de seguro, conseguimos codificar os pontos em cores por `'smoker'`, e plotar as outras duas colunas (`'bmi', 'charges'`) nos eixos

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker']);
```

O *scatterplot* mostra que enquanto nÃ£o fumantes tendem a pagar um pouco mais com IMC alto, fumantes pagam muito mais.

Para enfatizar ainda mais esse fato, podemos usar o comando *sns.lmplot* para adicionar duas linhas de regressÃ£o, correspondete aos fumantes e nÃ£o fumantes:

```python
sns.lmplot(x='bmi', y='charges', hue='smoker', data=insurance_data);
```

ðŸ‘€ <font color='white'>**NOTE:**</font> A linha de regressÃ£o para fumantes tem uma inclinaÃ§Ã£o muito mais acentuada em relaÃ§Ã£o a linha para nÃ£o fumantes!

Exite mais um grÃ¡fico que parece um pouco diferente de como vocÃª estÃ¡ acostumado a ver um *scatterplot*.

Normalmente, usamos o *scatterplot* para destacar a relaÃ§Ã£o entre duas variÃ¡veis contÃ­nuas. Entretanto podemos adaptar o *design* do *scatterplot* para apresentar uma variÃ¡vel categÃ³rica (como *'smoker'*) em um dos eixos principais.

Vamos referir a esse tipo de grÃ¡fico como um grÃ¡fico de dispersÃ£o categÃ³rica e construÃ­-lo com o comando `sns.swarmplot`:

```python
plt.figure(figsize=(13, 10))
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges']);
```

Duas informaÃ§Ãµes que o grÃ¡fico nos trÃ¡s:
- em mÃ©dia, os nÃ£o fumantes pagam menos que os fumantes
- os clientes que pagam mais sÃ£o fumantes; enquanto os clientes que pagam menos sÃ£o os nÃ£o fumantes

## Distributions

### Set up

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

### Load the data 

```python
iris_data = pd.read_csv(r'C:\Users\migsk\Documents\Kaggle Courses\kaggle-courses\Data Visualization Course\interesting-data-to-visualize\iris.csv', low_memory=False, index_col='Id')
iris_data
```

### Histograms

Vamos criar um grÃ¡fico para ver como o comprimento das pÃ©talas varia nas flores de Ã­ris:

```python
sns.histplot(iris_data['Petal Length (cm)']);
```

### Density plots

O prÃ³ximo tipo de plot Ã© o **kernel density estimate (KDE)**. Podemos pensar nele como um histograma mais suave.

```python
sns.kdeplot(data=iris_data['Petal Length (cm)'], fill=True);
```

### 2D KDE plots

NÃ£o estamos restritos a uma Ãºnica coluna quando criamos um grÃ¡fico KDE. Podemos criar um grÃ¡fico KDE 2D com o comando `sns.jointplot`.

```python
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind='kde');
```

Observe que, alÃ©m do grÃ¡fico 2D do KDE no centro,
- a curva no topo da figura Ã© um grÃ¡fico KDE para os dados do eixo x (iris_data['Petal Lenght (cm)']) e
- a curva na direita da figura Ã© um grÃ¡fico KDE para os dados do eixo y (iris_data['Sepal Width (cm)])

### Color-codded plots

NÃ³s conseguimos criar 3 histogramas de comprimento da pÃ©tala diferentes (um para cada espÃ©cie):

```python
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species');

plt.title('Histogram of Petal Lengths, by Species');
```

- `data=` fornece o nome da variÃ¡vel que usaremos para ler os dados
- `x=` define o nome da coluna que estÃ¡ os dados que queremos plotar
- `hue=` define a coluna que usaremos para dividir os dados em diferentes histogramas

NÃ³s tambÃ©m podemos criar um grÃ¡fico KDE para cada espÃ©cie usando `sns.kdeplot`. A funcionalidade para `data`, `x`, `hue` sÃ£o identicas quando usamos o `sns.histplot`. Adicionalmente, definimos `fill=True` para colorir a Ã¡rea abaixo da curva.

```python
sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', fill=True)

plt.title('Distribution of Petal Lengths, by Species');
```

ðŸ‘€ **PEGA VISÃƒO:**

Um padrÃ£o interessante que conseguimos ver nos grÃ¡ficos Ã© que as plantas aparentam pertencem a um dos dois grupos, onde *iris-versicolor* e *iris-virginica* parecem ter valores similares para comprimento da pÃ©tala, onde *iris-setosa* pertence a uma categoria prÃ³pria.

De fato, de acordo com o *dataset*, podemos atÃ© classificar qualquer planta como *iris-setosa* apenas olhando para o comprimento da pÃ©tala: se o comprimento Ã© menor que 2cm, Ã© mais provÃ¡vel de ser uma *iris-setosa*, ao contrÃ¡rio das outras duas espÃ©cies que no intervalo de comprimento 4-6cm precisaria de outras anÃ¡lises para definir qual Ã© a espÃ©cie.

## Chossing Plot Types and Custom Styles

### Chossing Plot Types

- **TendÃªncias:** Ã© definida como uma mudanÃ§a de padrÃ£o.
    - `sns.lineplot` - **line charts** sÃ£o os melhores para mostrar tendÃªncias em um perÃ­odo de tempo e mÃºltiplas linhas pode ser usadas para mostrar a tendÃªncia de mais de um grupo.

- **RelaÃ§Ã£o:** existe muitos tipos diferentes de grÃ¡ficos que vocÃª pode usar para visualizar relaÃ§Ãµes entre variÃ¡veis em seus dados.
    - `sns.barplot` - **bar charts** sÃ£o Ãºteis para comparar quantidades correspondente a dois grupos diferentes.
    - `sns.heatmap` - **heatmaps** podem ser usados para encontrar padrÃµes de cores codificadas em tabelas de nÃºmeros.
    - `sns.scatterplot` - **scatter plot** mostra a relaÃ§Ã£o entre duas variÃ¡veis contÃ­nuas; se codificada em cores, podemos tambÃ©m mostrar a relaÃ§Ã£o com uma terceira variÃ¡vel categorica.
    - `sns.regplot` - inclui uma **linha de regressÃ£o** em um *scatter plot* torna mais fÃ¡cil para visualizar qualquer relaÃ§Ã£o linear entre as duas variÃ¡veis.
    - `sns.lmplot` - desenha mÃºltiplas linhas de regressÃ£o, se o *scatter plot* tiver mais de uma variÃ¡vel e codificado em cor.
    - `sns.swarmplot` - **Categorical scatter plot** mostra a relaÃ§Ã£o entre variÃ¡veis continuas e categoricas.

- **DistribuiÃ§Ã£o:** Visualizamos distribuiÃ§Ã£o para mostrar possÃ­vel valores que podemos esperar ver em uma variÃ¡vel, juntamente com a probabilidade de eles serem.
    - `sns.histplot` - **Histogram** mostra a distribuiÃ§Ã£o de uma Ãºnica variÃ¡vel numÃ©rica.
    - `sns.kdsplor` - **KDE plots (or 2D KDE pllots)** mostram uma estimativa, distribuiÃ§Ã£o suave de uma Ãºnica variÃ¡vel numÃ©rica (ou duas, no caso do 2D).
    - `sns.jointplot` - Ãºtil para mostrar simultaneamente um 2D KDE plot com os correspondentes KDE plots para cada variÃ¡vel individual.

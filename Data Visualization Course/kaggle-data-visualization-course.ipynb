## Bar Charts and Heatmaps

### Set up

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

### Load the data

```python
fligth_data = pd.read_csv(r'C:\Users\migsk\Documents\Kaggle Courses\kaggle-courses\Data Visualization Course\interesting-data-to-visualize\flight_delays.csv', low_memory=False, index_col="Month")
```

```python
fligth_data
```

### Bar Chart

Vamos criar um gráfico de barras mostrando o atraso médio de chegada dos voos da Spirit Airlines *(código da companhia aérea: NK)* por mês:

```python
# Definindo o tamanho da figura
plt.figure(figsize=(16, 4))

# Adicionando o título do gráfico
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')

# Gráfico de barras
sns.barplot(x=fligth_data.index, y=fligth_data['NK'])

# Adicionando rótulo para o eixo vertical
plt.ylabel('Arrival delay (in minutes)');
```

> 👀 <font color='white'>**NOTE:**</font> Selecionamos a coluna de índice assim: flight_data.index, não seria possível usar flight_data['Month']. Isso acontece porque quando carregamos o dataset, usamos a coluna 'Month' como índice.

### Heatmap

Vamos criar um *heatmap* para vizualizar padrões em *flight_data*. Cada célula é codificada por cores de acordo com seu valor correspondente:

```python
# Definindo o tamanho da figura
plt.figure(figsize=(14, 7))

# Adicionando um título
plt.title('Average Arrival Delay for Each Airline, by Month')

# Criando o heatmap
sns.heatmap(data=fligth_data, annot=True) # Parâmetro 'annot' é para mostrar os valores em cada célula

# Adiciona um rótulo para o eixo horizontal
plt.xlabel('Airline');
```

🔍 Com esse gráfico podemos observar que os meses do final do ano (especialmente os meses 9-11) parecem relativamente escuros para todas as companhias aéreas. Isso sugere que as companhias aéreas são melhores (em média) em manter o cronograma durante esses meses!

## Scatter Plots

### Set up

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

### Load the data

```python
insurance_data = pd.read_csv(r'C:\Users\migsk\Documents\Kaggle Courses\kaggle-courses\Data Visualization Course\interesting-data-to-visualize\insurance.csv', low_memory=False)
```

```python
insurance_data
```

### Scatter Plots

Para criar um **scatter pllot** simples, nós usamos o comando *sns.scatterplot*, especificando os valores para:
 - o eixo horizontal (x)
 - o eixo vertical (y)

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges']);
```

O **scatterplot** acima sugere que *body mass index - BMI* (índice de massa corporal - IMC) e *insurance charges* (taxa de seguro) são <font colro='LimeGreen'>**positivamente correlacionado**</font>, onde clientes com maiores BMI normalmente também tendem a pagar mais em custos de seguro.

> Esse padrão faz sentido, já que o IMC alto costuma estar associado a um risco maior de doenças crônicas

Para uma checagem mais reforçada nessa relação, você deve adicionar uma **linha de regressão**, ou a melhor linha que encaixar nos dados. Nós fazemos isso com o comando *sns.regplot*.

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges']);
```

### Color-codded scatter plots

Nós conseguimos usar o *scatter plot* para mostrar a relação entre três variáveis! Uma maneira de fazer isso é codificar os pontos em cores.

Por exemplo, para entender como fumar afeta a relação entre IMC e a taxa de seguro, conseguimos codificar os pontos em cores por `'smoker'`, e plotar as outras duas colunas (`'bmi', 'charges'`) nos eixos

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker']);
```

O *scatterplot* mostra que enquanto não fumantes tendem a pagar um pouco mais com IMC alto, fumantes pagam muito mais.

Para enfatizar ainda mais esse fato, podemos usar o comando *sns.lmplot* para adicionar duas linhas de regressão, correspondete aos fumantes e não fumantes:

```python
sns.lmplot(x='bmi', y='charges', hue='smoker', data=insurance_data);
```

👀 <font color='white'>**NOTE:**</font> A linha de regressão para fumantes tem uma inclinação muito mais acentuada em relação a linha para não fumantes!

Exite mais um gráfico que parece um pouco diferente de como você está acostumado a ver um *scatterplot*.

Normalmente, usamos o *scatterplot* para destacar a relação entre duas variáveis contínuas. Entretanto podemos adaptar o *design* do *scatterplot* para apresentar uma variável categórica (como *'smoker'*) em um dos eixos principais.

Vamos referir a esse tipo de gráfico como um gráfico de dispersão categórica e construí-lo com o comando `sns.swarmplot`:

```python
plt.figure(figsize=(13, 10))
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges']);
```

Duas informações que o gráfico nos trás:
- em média, os não fumantes pagam menos que os fumantes
- os clientes que pagam mais são fumantes; enquanto os clientes que pagam menos são os não fumantes

## Distributions

### Set up

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

### Load the data 

```python
iris_data = pd.read_csv(r'C:\Users\migsk\Documents\Kaggle Courses\kaggle-courses\Data Visualization Course\interesting-data-to-visualize\iris.csv', low_memory=False, index_col='Id')
iris_data
```

### Histograms

Vamos criar um gráfico para ver como o comprimento das pétalas varia nas flores de íris:

```python
sns.histplot(iris_data['Petal Length (cm)']);
```

### Density plots

O próximo tipo de plot é o **kernel density estimate (KDE)**. Podemos pensar nele como um histograma mais suave.

```python
sns.kdeplot(data=iris_data['Petal Length (cm)'], fill=True);
```

### 2D KDE plots

Não estamos restritos a uma única coluna quando criamos um gráfico KDE. Podemos criar um gráfico KDE 2D com o comando `sns.jointplot`.

```python
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind='kde');
```

Observe que, além do gráfico 2D do KDE no centro,
- a curva no topo da figura é um gráfico KDE para os dados do eixo x (iris_data['Petal Lenght (cm)']) e
- a curva na direita da figura é um gráfico KDE para os dados do eixo y (iris_data['Sepal Width (cm)])

### Color-codded plots

Nós conseguimos criar 3 histogramas de comprimento da pétala diferentes (um para cada espécie):

```python
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species');

plt.title('Histogram of Petal Lengths, by Species');
```

- `data=` fornece o nome da variável que usaremos para ler os dados
- `x=` define o nome da coluna que está os dados que queremos plotar
- `hue=` define a coluna que usaremos para dividir os dados em diferentes histogramas

Nós também podemos criar um gráfico KDE para cada espécie usando `sns.kdeplot`. A funcionalidade para `data`, `x`, `hue` são identicas quando usamos o `sns.histplot`. Adicionalmente, definimos `fill=True` para colorir a área abaixo da curva.

```python
sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', fill=True)

plt.title('Distribution of Petal Lengths, by Species');
```

👀 **PEGA VISÃO:**

Um padrão interessante que conseguimos ver nos gráficos é que as plantas aparentam pertencem a um dos dois grupos, onde *iris-versicolor* e *iris-virginica* parecem ter valores similares para comprimento da pétala, onde *iris-setosa* pertence a uma categoria própria.

De fato, de acordo com o *dataset*, podemos até classificar qualquer planta como *iris-setosa* apenas olhando para o comprimento da pétala: se o comprimento é menor que 2cm, é mais provável de ser uma *iris-setosa*, ao contrário das outras duas espécies que no intervalo de comprimento 4-6cm precisaria de outras análises para definir qual é a espécie.

## Chossing Plot Types and Custom Styles

### Chossing Plot Types

- **Tendências:** é definida como uma mudança de padrão.
    - `sns.lineplot` - **line charts** são os melhores para mostrar tendências em um período de tempo e múltiplas linhas pode ser usadas para mostrar a tendência de mais de um grupo.

- **Relação:** existe muitos tipos diferentes de gráficos que você pode usar para visualizar relações entre variáveis em seus dados.
    - `sns.barplot` - **bar charts** são úteis para comparar quantidades correspondente a dois grupos diferentes.
    - `sns.heatmap` - **heatmaps** podem ser usados para encontrar padrões de cores codificadas em tabelas de números.
    - `sns.scatterplot` - **scatter plot** mostra a relação entre duas variáveis contínuas; se codificada em cores, podemos também mostrar a relação com uma terceira variável categorica.
    - `sns.regplot` - inclui uma **linha de regressão** em um *scatter plot* torna mais fácil para visualizar qualquer relação linear entre as duas variáveis.
    - `sns.lmplot` - desenha múltiplas linhas de regressão, se o *scatter plot* tiver mais de uma variável e codificado em cor.
    - `sns.swarmplot` - **Categorical scatter plot** mostra a relação entre variáveis continuas e categoricas.

- **Distribuição:** Visualizamos distribuição para mostrar possível valores que podemos esperar ver em uma variável, juntamente com a probabilidade de eles serem.
    - `sns.histplot` - **Histogram** mostra a distribuição de uma única variável numérica.
    - `sns.kdsplor` - **KDE plots (or 2D KDE pllots)** mostram uma estimativa, distribuição suave de uma única variável numérica (ou duas, no caso do 2D).
    - `sns.jointplot` - útil para mostrar simultaneamente um 2D KDE plot com os correspondentes KDE plots para cada variável individual.
